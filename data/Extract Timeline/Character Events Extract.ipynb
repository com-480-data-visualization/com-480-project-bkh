```python
names = ["Gandalf","Frodo","Aragorn","Bilbo","Gollum"]
def find_names (row):
    chars = ""
    for name in names:
        if name in row["Event"]:
            chars = chars + "," + name
    if chars == "":
        return None
    return chars[1:]
```

```python
df["Names"] = df.apply (lambda row: find_names(row), axis=1)
```

```python
df_names = df.dropna()
```

```python
new_df = new_df.rename(columns={0: "Name"})
```
