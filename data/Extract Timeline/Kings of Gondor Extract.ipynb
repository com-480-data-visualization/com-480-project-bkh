```python
import requests

URL = 'https://lotr.fandom.com/wiki/Kings_of_Gondor'
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
tables  = soup.findAll("table")
```

```python
len(tables)
```

```python
d = []
for row in tables[1].findAll("tr")[1:]:
    col = row.findAll("td")[1:]
    name = "Dummy"
    try:
        counter = 0
        for cc in col[0].findAll("a"):
            name = cc.get("title")
            if name is not None :
                break
        begin = col[1].findAll("a")[0].get("title")
        end = col[1].findAll("a")[1].get("title")
        era = begin[0:2]
        begin = int(begin[3:])
        end = int(end[3:])
        begin
        if era == "SA":
            era = 0
        elif era == "TA":
            era = 1
        else:
            era =2
        d.append(
            {
                "id":name,
                "start":begin,
                "end":end,
                "lane":era
            }
        )
    except:
        print("cant parse")
        d.append(
            {
                "id":"dumb",
                "start":0,
                "end":0,
                "lane":-1
            }
        )
```
